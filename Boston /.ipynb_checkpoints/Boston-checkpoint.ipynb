```python
def GD(x,y,learning_rate,no_of_iterations):
    m=np.zeros(len(x[0]))
    for i in range(no_of_iterations):
        m=step_gradient(x, y, m, learning_rate)
        print("itr= ", i, "cost=", end=' ')
        Cost(m, x, y)
    return m
```

```python
def Cost(m,x,y):
    cost=0
    n=len(x)
    for i in range(n):
        cost+= (1/n)*((y[i]-((x[i]*m).sum()))**2)
    print(cost)
```

```python
def step_gradient(X, Y, m, learning_rate):
    m_slope=np.zeros(len(X[0]))
    for i in range(len(X)):
        x=X[i]
        y=Y[i]
        for j in range(len(x)):
            m_slope[j]+=(-2/len(X))*(y-sum(m*x))*x[j]
    new_m=m-(learning_rate*m_slope)
    return new_m
```

```python
def Gradient_Descent(x,y):
    iterations=300
    learning_rate=0.1
    x=np.append(x, np.ones(len(x)).reshape(-1, 1), axis=1)
    m=GD(x, y, learning_rate, iterations)
    return m
```

```python
x=training[:, :-1]
y=training[:, -1]
#adding squared values of each column
sq=[]
for i in x:
    sq.append(i**2)
sq=np.array(sq)
x=np.append(x, sq, axis=1)

scaler=preprocessing.StandardScaler()
scaler.fit(x)
x=scaler.transform(x)
m=Gradient_Descent(x, y)
```

```python
sq=[]
for i in testing:
    sq.append(i**2)
sq=np.array(sq)
testing=np.append(testing, sq, axis=1)

testing1=scaler.transform(testing)
x_test=np.append(testing1, np.ones(len(testing1)).reshape(-1, 1), axis=1)
ans=[]
for i in x_test:
    ans.append(sum(i*m))
for i in ans:
    print(i)
ans=np.array(ans)
np.savetxt(X=ans,fname='results.csv' , delimiter=',', fmt='%.5f')
```
